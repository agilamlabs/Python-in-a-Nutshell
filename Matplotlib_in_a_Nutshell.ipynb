##### markersize or the shorter ms

##### markeredgecolor or the shorter mec

##### markerfacecolor or the shorter mfc

```python
import matplotlib.pyplot as plt
import numpy as np

y1 = np.array([3, 8, 1, 10])
y2 = np.array([6, 2, 7, 8])

plt.plot(y1, marker = 'o',ms = 20, mec = 'r')
plt.plot(y2, marker = 'o', ms = 10, mfc = 'g')

plt.show()
```

##### linestyle can be written as ls

##### color or the shorter c

##### linewidth or the shorter lw

##### dotted can be written as :

##### dashed can be written as --

```python
import matplotlib.pyplot as plt
import numpy as np

y1 = np.array([3, 8, 1, 10])
y2 = np.array([6, 2, 7, 8])

plt.plot(y1, ls = 'dashed', c = 'y')
plt.plot(y2, lw = '5')

plt.show()
```

## Multiple Plots

```python
"""
plt.subplot(1, 2, 1)
the figure has 1 row, 2 columns, and this plot is the first plot.

plt.subplot(1, 2, 2)
the figure has 1 row, 2 columns, and this plot is the second plot. 

"""
```

```python
import matplotlib.pyplot as plt
import numpy as np

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)   #1 row, 2 column, 1st plot
plt.title("Sales")     #1st plot Title
plt.plot(x,y)


#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)   #1 row, 2 column, 2nd plot
plt.title("income")    #2nd plot Title
plt.plot(x,y)

plt.suptitle("shop")  #Super Title
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 1, 1)   #2 row, 1 column, 1st plot
plt.title("sales")    #1st plot Title
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 1, 2)   #2 row, 1 column, 2nd plot
plt.title("income")    #2nd plot Title
plt.plot(x,y)

plt.suptitle("Shop")   #Super Title
plt.show()
```

## Bar graphs

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.bar(x,y, color = "red")    #bar graph

plt.subplot(1, 2, 2)
plt.barh(x, y, color = "#4CAF50")  #bar in horizontal 

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.bar(x, y,color = "red", width = 0.1)   #width in vertical

plt.subplot(1, 2, 2)
plt.barh(x, y,color = "#4CAF50", height = 0.1) #height in horizontal

plt.show()
```
